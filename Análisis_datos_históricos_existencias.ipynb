## Análisis de los datos históricos de ingresos y Creación de un Cuadro de Mando

```python
import yfinance as yf
import pandas as pd
import matplotlib_inline
import matplotlib.pyplot as plt
```

## Question 1: Use the yfinance to Extract Stock Data


```python
# Crea un objeto para Tesla con el símbolo de ticker 'TSLA'
tesla = yf.Ticker("TSLA")

# Obtén los datos históricos de Tesla con el periodo máximo
tesla_data = tesla.history(period="max")

# Imprime las primeras filas para verificar
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

## Question 2: Use Webscraping to Extract Tesla revenue Data

```python
# Crea un objeto para Tesla con el símbolo de ticker 'TSLA'
tesla = yf.Ticker("TSLA")

# Obtén los datos financieros de Tesla
tesla_financials = tesla.financials.T

# Convierte los datos financieros a un DataFrame
tesla_revenue = tesla_financials[['Total Revenue']].reset_index()
tesla_revenue.columns = ['Date', 'Revenue']

# Imprime las últimas filas para verificar
tesla_revenue.tail()
```

## Question 5: Tesla Stock Price History

```python
# Función para crear el gráfico
def make_graph(data, title):
    plt.figure(figsize=(8, 4))
    plt.plot(data.index, data['Close'], label='Close Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    plt.grid(True)
    plt.show()

# Llama a la función make_graph con los datos de Tesla y un título
make_graph(tesla_data, 'Tesla Stock Price History')
```
